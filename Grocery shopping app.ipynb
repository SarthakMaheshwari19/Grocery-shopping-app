```python
inventory=[] 
delivery_charges={5: 10.0, 10:20, 15: 30.0}

def add_item_to_inventory():

    serial_no=len(inventory)+1  
    item_name=input("Enter the name of the item: ")
    quantity=int(input("Enter the quantity: "))
    price=float(input("Enter the price per item: "))

    for i, item in enumerate(inventory):
        if item[1]==item_name:
            inventory[i][2] +=quantity
            break
    else: 
        inventory.append([serial_no, item_name, quantity, price])

def display_inventory():
    
    print("\nInventory:")
    print("Ser.No.\tName\tQuantity\tPrice ") 
    print("-"*40)
    
    for item in inventory:
        print("%d\t%-10s\t%d\t\t%.2f" % tuple(item)) 
    
    print("-"*40)

def calculate_total_cost(order):
    
    total = 0
    for item, qty in order:
        total += qty * item[3]
    return total
    
def calculate_delivery_charge(distance):
  
    return delivery_charges.get(distance, 0)

def place_order():

    order=[] 
    name=input("\nEnter customer name: ")
    address=input("Enter address: ")

    display_inventory()
    
    while True:
        serial_no = int(input("\nEnter serial no (0 to exit): "))   
        if serial_no == 0:
            break
            
        if serial_no <=len(inventory) and serial_no > 0: 
            
            item=inventory[serial_no-1] 
            available_quantity=item[2]
            ordered_quantity=int(input("Enter quantity: "))

            if ordered_quantity<=available_quantity:
                order.append((item, ordered_quantity))
            else:
                print(f"Sorry, only {available_quantity} units of {item[1]} available. Try again with a lower quantity.")
        else:
            print("Invalid serial number. Try again.")

    distance=int(input("\nEnter delivery distance (5/10/15 km): "))  

    delivery_charge=calculate_delivery_charge(distance)
    total_item_cost= calculate_total_cost(order)
    total_cost= total_item_cost + delivery_charge

    print("\nBILL")
    print("-"*30)
    print("%-20s: %s" % ("Customer Name", name))
    print("%-20s: %s" % ("Customer Address", address))

    print("\nBill Details:")
    
    print("="*30)
    print("%-10s %-10s %-10s" % ("Item", "Quantity", "Price"))
    print("="*30)
    
    for item, qty in order: 
        name = item[1]
        price = item[3]
        print("%-10s %-10d %-10.2f" % (name, qty, price))

    print("="*30)

    print("%-10s %-10.2f" % ("Total Item Cost", total_item_cost)) 
    print("%-10s %-10.2f" % ("Delivery Charge", delivery_charge))
    print("%-10s %-10.2f" % ("Total Cost", total_cost))
    print("="*30)

    # Updating inventory  
    for item, qty in order:
        item[2] -= qty 

```

```python
def main():
    while True:
        choice = int(input("""Enter your choice (1-3, 0 to exit):\n
                           1: To add item to inventory\n
                           2: To display inventory\n
                           3: To place order\n"""))
        if choice == 0:
            print("Thank you for shopping with us")
            break
        elif choice == 1:
            add_item_to_inventory()
        elif choice == 2:
            display_inventory()
        elif choice == 3:
            place_order()
        else:
            print("Invalid choice. Please choose from 1 to 3.")

if __name__ == "__main__":
    main()
```
